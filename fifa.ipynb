## Imputation the missing values

```python
for i in fifa .columns:
    if fifa[i].isnull().any():
        if fifa[i].dtypes=="object":
            fifa[i].fillna(fifa[i].mode()[0],inplace=True)
        elif  fifa[i].dtypes=="float":
            fifa[i].fillna(fifa[i].median(),inplace=True)
        elif  fifa[i].dtypes=="int":
            fifa[i].fillna(fifa[i].median(),inplace=True)
        else:
            pass
```
